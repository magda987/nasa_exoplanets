```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
url = "http://127.0.0.1:8000/plmass?limit=10000"
data = pd.read_json(url)
data
```

```python
data = data.dropna()
data.count()
data.describe()
```

```python
x = data.drop(columns = ["pl_masse"]) # x is temp, so drop mass
y = data["pl_masse"]
y
```

```python
plt.xlabel("star temperature")
plt.ylabel("planet mass")
plt.scatter(x,y)
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2960],[5000]])
```

```python
plt.plot(x, model.predict(x), color = "red")
```

```python
r_sq = model.score(x,y)
r_sq
```
